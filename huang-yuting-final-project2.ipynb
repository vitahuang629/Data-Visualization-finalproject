```python
##GRID HEAT MAP##

# 2. scales
col_sc = bqplot.ColorScale(scheme='Purples')
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

# 3. axis
col_ax = bqplot.ColorAxis(scale=col_sc, orientation='horizontal', side='top', label = 'log10 Color Scale: ')
x_ax = bqplot.Axis(scale=x_sc, label='Stratification',tick_rotate=320, tick_style={'font-size': '14px'}, label_offset ='50px')

y_ax = bqplot.Axis(scale=y_sc, orientation='vertical', label='Location',tick_offset = '50px',tick_style ={'font-size':'14px'}, label_offset ='50px')

# 4. marks
heat_map = bqplot.GridHeatMap(color = np.log10(table.values),
                             row = location,
                             column = Stratification, 
                             scales = {'color':col_sc,'row':y_sc, 'column':x_sc},
                             interactions={'click':'select'},
                             selected_style={'fill':'pink'})

fig = bqplot.Figure(marks=[heat_map], axes=[col_ax,x_ax, y_ax])
fig.layout.height = '800px'
```

```python
def on_selection(change):
    if len(change['owner'].selected==1): # only allow user to select one grid
        i,j = change['owner'].selected[0]
        v = table.values[i,j]
        #myLabel = myLabel(color = '#ff0000')
        myLabel.value = 'Total sample size for ' +  Stratification[j] +' class '+'in '+ location[i] +' is: ' +str(v) +' ft^2'
        
        new_location = table.index[i]
        new_category = table.columns[j]
        filter = (nutrition["LocationAbbr"] == new_location) & (nutrition["StratificationCategory1"] == new_category) & (nutrition["Question"] == "Percent of adults aged 18 years and older who have an overweight classification")
        subset = nutrition[filter]
        
        strata_type = subset[new_category].values
        percentage = subset["Data_Value"].values
        
        bar.x = strata_type
        bar.y = percentage
```

```python
heat_map.observe(on_selection,'selected')
```

```python
fig.layout.min_width='300px'
fig.layout.min_height='300px'
myDashboard = ipywidgets.VBox([myLabel,fig])
```

```python
# Scales
x_scb = bqplot.OrdinalScale()
y_scb = bqplot.LinearScale()

# Axis
x_axb = bqplot.Axis(label='Stratification Type', scale=x_scb, tick_rotate=330)
y_axb = bqplot.Axis(label="Percentage of adults aged 18+ and classified as obesity", scale=y_scb, orientation='vertical')

i, j = 0, 0
new_location = table.index[i]
new_stratification = table.columns[j]

# mask
mask = nutrition[(nutrition["LocationAbbr"] == new_location) & (nutrition["StratificationCategory1"] == new_stratification) & (nutrition["Question"] == "Percent of adults aged 18 years and older who have an overweight classification")]
strata_type = mask[new_stratification].values
percentage = mask["Data_Value"].values

# Marks
bar = bqplot.Bars(x=strata_type,
                  y=percentage,
                  scales={"x": x_scb, "y": y_scb})
                  
fig_bar = bqplot.Figure(marks=[bar], axes=[x_axb, y_axb], padding_x=0.025, padding_y=0.025)
fig_bar
```

```python
# Reference: https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_with_legend.html
def nutrition_xy(x, y):
    with plt.style.context('seaborn'):
        #fig = plt.figure(figsize=(16, 10))
        fig, ax = plt.subplots(figsize=(16, 10))
        scatter = ax.scatter(x = nutrition[x], y = nutrition[y],c =np.log10(nutrition['Sample_Size'])  ,s = 20, cmap = 'Accent')
        legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Sample_Size")
        #plt.scatter(x = nutrition[x], y = nutrition[y],c =np.log10(nutrition['Sample_Size'])  ,s = 20,cmap = 'rainbow')
        plt.xlabel(x.capitalize())
        plt.ylabel(y.capitalize())
        plt.title("%s vs %s"%(x.capitalize(), y.capitalize()))
        plt.show()
```

```python
out = ipywidgets.interactive_output(nutrition_xy, {'x': x, 'y':y})
```

```python
ipywidgets.VBox([ipywidgets.HBox([x, y]), out])
```
